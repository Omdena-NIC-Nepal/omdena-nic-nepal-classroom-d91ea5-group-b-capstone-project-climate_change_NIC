```python
import pandas as pd
import numpy as np
```

```python
ag_data = pd.read_csv("X:/AI DataScience projects/Assignment 8/capstone-project-KarunBhandari/Weather_&_Climate_Data/Historical temperature, precipitation, and weather events/raw/agricultural_data/agriculture-and-rural-development_npl.csv")
#ag_data.describe()
ag_data.info()
ag_data.head()
```

```python

pivoted_df = ag_data.pivot_table(
    index=['Country Name', 'Country ISO3', 'Year'],
    columns='Indicator Name',
    values='Value',
    aggfunc='first'  
).reset_index()

pivoted_df.columns = [col.replace(' ', '_').replace('(', '').replace(')', '').replace('%', 'pct') 
                     for col in pivoted_df.columns]

pivoted_df.columns = pivoted_df.columns.str.strip()
pivoted_df.head()


#pivoted_df.to_csv('X:/AI DataScience projects/Assignment 8/capstone-project-KarunBhandari/Weather & Climate Data/Processed Datas/transformed_agriculture_data.csv', index=True)
```
